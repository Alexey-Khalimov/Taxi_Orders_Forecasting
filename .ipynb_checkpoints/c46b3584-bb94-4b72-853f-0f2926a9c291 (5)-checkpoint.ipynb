#  Прогнозирование заказов такси

**ПОСТАНОВКА ЗАДАЧИ**

**Заказчик исследования:**  
Компания «Чётенькое такси»

**Исходные данные:**  
Исторические данные о количестве заказов такси в аэропортах, предоставленные в файле `taxi.csv`. Данные содержат временной ряд с метками времени и количеством заказов за каждый час. Целевой признак — количество заказов, который находится в столбце `num_orders`.

**Цель исследования:**  
Построить модель для прогнозирования количества заказов такси на следующий час, которая поможет компании привлекать больше водителей в периоды пиковой нагрузки. Целевое значение метрики RMSE на тестовой выборке не должно превышать 48.

**Задачи исследования:**
1. Загрузить и обработать данные, выполненные ресемплированием по одному часу.
2. Провести исследовательский анализ данных (EDA), чтобы выявить тренды, сезонность, аномалии и другие важные особенности временного ряда.
3. Подготовить данные для моделирования, включая создание новых признаков, добавление временных меток и других дополнительных переменных, таких как праздничные дни, выходные, время суток и т.д.
4. Разделить данные на обучающую и тестовую выборки, выделив 10% данных для тестирования.
5. Обучить несколько моделей для прогнозирования временного ряда, включая Prophet, CatBoost, Naive и другие, настроив различные гиперпараметры.
6. Оценить качество каждой модели на тестовой выборке с использованием метрики RMSE.
7. Сделать выводы о качестве работы моделей и предложить рекомендации по использованию наилучшей модели.

**Ход исследования:**
1. **Загрузка и подготовка данных:** Загрузка данных из файла, приведение данных к часовому формату с помощью ресемплирования и первичная очистка (обработка пропусков и выбросов).
   
   
2. **Исследовательский анализ данных (EDA):**  
   - Анализ временных рядов для поиска трендов и сезонных паттернов.
   - Выявление влияния различных факторов (день недели, время суток, праздники и т.д.) на количество заказов.
   
   
3. **Создание признаков:**  
   - Добавление временных признаков: час, день недели, выходные, праздники и преобразование циклических признаков (синус/косинус).
   - Построение лагов и скользящих средних для учёта зависимости заказов от предыдущих значений.
   
   
4. **Разделение данных:** Разделение данных на обучающую и тестовую выборки, где тестовая выборка составит 10% от общего объёма данных.


5. **Обучение моделей:**  
   - Обучение моделей Prophet, CatBoost, Naive, а также настройка гиперпараметров для каждой модели.
   - Применение бэктестинга для оценки точности прогнозов на различных временных отрезках.


6. **Оценка качества моделей:**  
   - Оценка каждой модели на тестовой выборке с использованием метрики RMSE.
   - Сравнение результатов разных моделей и выбор наилучшей по качеству прогнозов.


7. **Выводы и рекомендации:**  
   - Анализ качества каждой модели, выбор наилучшей модели для прогнозирования.
   - Оценка влияния различных признаков и предложений по дальнейшему улучшению моделей.

```python
# Функция для загрузки данных
def load_data(local_file_path, remote_url, sep=',', decimal='.', date_columns=None, index_col=None):
    try:
        if os.path.exists(local_file_path):
            print(f'Загрузка данных из локального файла: {local_file_path}')
            return pd.read_csv(local_file_path, sep=sep, decimal=decimal, parse_dates=date_columns, index_col=index_col)
        else:
            print(f'Загрузка данных из удаленного источника: {remote_url}')
            return pd.read_csv(remote_url, sep=sep, decimal=decimal, parse_dates=date_columns, index_col=index_col)
    except Exception as e:
        # Если возникла ошибка, выводим сообщение об ошибке
        print('Загрузка произошла с ошибкой')
        print(f'Ошибка: {str(e)}')
        return None
```

```python
# Функция для первичного обзора данных
def data_overview(df, date_columns=None):
    """
    Функция для обзора данных, включает:
    1. Вывод первых 5 строк
    2. Общую информацию о данных
    3. Проверку пропусков (абсолютные и относительные значения)
    4. Распределение количественных столбцов
    5. Уникальные значения для категориальных столбцов
    6. Количество уникальных значений в количественных данных
    7. Проверку на наличие дубликатов
    """
    # 1. Вывод первых 5 строк для быстрого обзора данных
    print("1. Первые 5 строк данных:")
    display(df.head())
    
    # Визуальное разделение
    print(LINE_1)
    
    # 2. Общая информация о данных: типы данных и количество ненулевых значений
    print("\n2. Информация о данных:")
    df.info()
    
    # Визуальное разделение
    print(LINE_1)
    
    # 3. Проверка пропусков: абсолютные и относительные значения (в процентах)
    print("\n3. Проверка пропусков в данных:")
    missing_abs = df.isnull().sum()  # Абсолютные значения пропусков
    missing_rel = (df.isnull().sum() / len(df)) * 100  # Пропуски в процентах
    missing_df = pd.DataFrame({'Пропуски (абсолютные)': missing_abs, 'Пропуски (%)': missing_rel})
    # Выводим информацию только о тех столбцах, где есть пропуски
    display(missing_df[missing_df['Пропуски (абсолютные)'] > 0])
    
    # Визуальное разделение
    print(LINE_1)
    
    # 4. Распределение количественных данных: ключевые статистические показатели
    print("\n4. Распределение количественных данных:")
    numeric_cols = df.select_dtypes(include=['int8', 'int16', 'int32', 'int64', 'float16', 'float32', 'float64']).columns
    display(df[numeric_cols].describe())  # Описание распределения (среднее, стандартное отклонение и т.д.)
    
    # Визуальное разделение
    print(LINE_1)
    
    # 5. Количество уникальных значений в количественных данных
    print("\n5. Количество уникальных значений в данных:")
    display(df.nunique())  # Выводим количество уникальных значений для числовых столбцов
        
    # Визуальное разделение
    print(LINE_1)
    
    # 6. Уникальные значения для категориальных столбцов: анализ категорий
    print("\n6. Уникальные значения для категориальных столбцов:")
    categorical_cols = df.select_dtypes(include=['object', 'category']).columns  # Выбираем только категориальные столбцы
    for col in categorical_cols:
        print(f"\nУникальные значения в столбце '{col}':")
        print(df[col].unique())  # Выводим уникальные значения для каждого категориального столбца
        
    # Визуальное разделение
    print(LINE_1)
    
    # 7. Проверка на дубликаты
    print("\n7. Проверка на наличие дубликатов:")
    duplicates_count = df.duplicated().sum()  # Считаем количество дубликатов
    if duplicates_count > 0:
        print(f"Найдено дубликатов: {duplicates_count}")
    else:
        print("Дубликатов не найдено.")

    # Визуальное разделение
    print(LINE_1)
    
    # 8. Проверка индексов на монотонность
    print("\n8. Проверка индексов на монотонность:", df.index.is_monotonic_increasing)        
    print("Минимальный и максимальный индекс:", df.index.min(), df.index.max())

    # Визуальное разделение
    print(LINE_1)
    
    # 9. Проверка колонок с временами на монотонность
    if date_columns:
        print("\n9. Проверка datetime на монотонность:", df[date_columns].index.is_monotonic_increasing)
        print("Минимальный и максимальный индекс:",  df[date_columns].min(), df[date_columns].max())
```

```python
# Функция для обзора временных рядов
def plot_time_series_decomposition(dataframe, column, model='additive', window=24, period=24, lags=[1, 2, 3, 4], 
                                   title_size=16, font_size=14, 
                                   plot_trend=True, plot_seasonality=True, plot_residuals=True,
                                   plot_moving_average=True, plot_acf_graph=True, plot_pacf_graph=True,
                                   plot_lag_scatter=True, plot_histogram=True):
    """
    Универсальная функция для отображения временного ряда с декомпозицией и дополнительными графиками
    :param df: DataFrame с временным рядом
    :param column: Имя столбца, содержащего временной ряд
    :param model: Тип модели декомпозиции временного ряда ('additive' или 'multiplicative')
    :param window: Размер окна для вычисления скользящего среднего
    :param period: Период для сезонной декомпозиции
    :param title_size: Размер шрифта для заголовков
    :param font_size: Размер шрифта для подписей осей и легенд
    :param plot_trend: Построить тренд
    :param plot_seasonality: Построить сезонность
    :param plot_residuals: Построить остатки
    :param plot_moving_average: Построить скользящее среднее
    :param plot_acf_graph: Построить автокорреляционную функцию (ACF)
    :param plot_pacf_graph: Построить частичную автокорреляционную функцию (PACF)
    :param plot_lag_scatter: Построить диаграммы рассеяния с лагами
    :param plot_rolling_std: Построить скользящее стандартное отклонение
    :param plot_histogram: Построить гистограмму распределения значений
    """
    df=dataframe.copy()
    
    # Считаем количество подграфиков для основной фигуры
    num_plots = sum([plot_moving_average, plot_trend, plot_seasonality, plot_residuals, plot_histogram])

    # Декомпозиция временного ряда
    decomposed = seasonal_decompose(df[column], model=model, period=period)

    # Вычисляем скользящее среднее
    df['moving_average'] = df[column].rolling(window=window, closed='left').mean()
    df['rolling_std'] = df[column].rolling(window=window, closed='left').std()  
    
    # Создаем фигуру и оси только для тех графиков, которые нужны
    if num_plots > 0:
        fig, axes = plt.subplots(num_plots, 1, figsize=(20, 6 * num_plots))
        if num_plots == 1:
            axes = [axes]  # Если один график, делаем его списком для универсальности кода
        idx = 0

        # 1. Исходные данные и скользящее среднее
        if plot_moving_average:
            axes[idx].plot(df.index, df[column], label='Исходные данные', alpha=0.7)
            axes[idx].plot(df.index, df['moving_average'], label=f'Скользящее среднее ({window} интервалов)', color='orange')
            axes[idx].plot(
                df.index, 
                df['rolling_std'], 
                label=f'Скользящее стандартное отклонение ({window} интервалов)', 
                color='red')
            axes[idx].set_title(
                f'Исходные данные, скользящее среднее и скользящее стандартное отклонение({window} интервалов)', 
                fontsize=title_size
            )
            axes[idx].set_xlabel('Дата и время', fontsize=font_size)
            axes[idx].set_ylabel(column, fontsize=font_size)
            axes[idx].legend(fontsize=font_size)
            axes[idx].grid(True, alpha=0.7)
            idx += 1

        # 2. Тренд
        if plot_trend:
            # Убираем NaN значения из тренда
            trend_notna = decomposed.trend.dropna()

            axes[idx].plot(trend_notna.index, trend_notna, label='Тренд', color='green')
            axes[idx].set_title('Тренд', fontsize=title_size)
            axes[idx].set_xlabel('Дата и время', fontsize=font_size)
            axes[idx].set_ylabel('Тренд', fontsize=font_size)
            axes[idx].grid(True, alpha=0.7)

            # Подготовка данных для аппроксимации трендов
            x = np.arange(len(trend_notna))  # Индексы временного ряда без NaN значений
            y = trend_notna.values  # Значения тренда без NaN

            # Линейная аппроксимация
            z1 = np.polyfit(x, y, 1)  # Параметры линейного тренда
            p1 = np.poly1d(z1)  # Полиномиальное представление тренда
            axes[idx].plot(trend_notna.index, p1(x), label='Линейный тренд', linestyle='--', color='blue')

            # Параболическая аппроксимация
            z2 = np.polyfit(x, y, 2)  # Параметры параболического тренда
            p2 = np.poly1d(z2)  # Полиномиальное представление тренда
            axes[idx].plot(trend_notna.index, p2(x), label='Параболический тренд', linestyle='-.', color='red')

            # Добавление легенды для различения трендов
            axes[idx].legend()
            idx += 1

        # 3. Сезонность
        if plot_seasonality:
            axes[idx].plot(decomposed.seasonal[:period].index, decomposed.seasonal[:period], label='Сезонность', color='purple')
            axes[idx].set_title(f'Сезонность за {period} интервалов', fontsize=title_size)
            axes[idx].set_xlabel('Дата и время', fontsize=font_size)
            axes[idx].set_ylabel('Сезонность', fontsize=font_size)
            axes[idx].grid(True, alpha=0.7)
            idx += 1

        # 4. Остатки
        if plot_residuals:
            axes[idx].plot(decomposed.resid.index, decomposed.resid, label='Остатки', color='lightcoral')
            axes[idx].set_title('Остатки', fontsize=title_size)
            axes[idx].set_xlabel('Дата и время', fontsize=font_size)
            axes[idx].set_ylabel('Остатки', fontsize=font_size)
            axes[idx].grid(True, alpha=0.7)
            idx += 1

        # 5. Гистограмма распределения значений
        if plot_histogram:
            sns.histplot(df[column], bins=30, kde=True, ax=axes[idx], color='skyblue', alpha=0.7)
            median_value = df[column].median()
            mean_value = df[column].mean()
            axes[idx].axvline(median_value, color='r', linestyle='--', label=f'Median: {median_value:.2f}')
            axes[idx].axvline(mean_value, color='g', linestyle='-', label=f'Mean: {mean_value:.2f}')
            axes[idx].legend(fontsize=title_size)
            axes[idx].set_title('Распределение значений временного ряда', fontsize=title_size)
            axes[idx].set_xlabel(column, fontsize=font_size)
            axes[idx].set_ylabel('Частота', fontsize=font_size)

        plt.tight_layout()
        plt.show()

    # 7. ACF (Автокорреляционная функция)
    if plot_acf_graph:
        fig, ax = plt.subplots(figsize=(20, 5))
        plot_acf(df[column], ax=ax, lags=period)
        ax.set_title('Автокорреляционная функция (ACF)', fontsize=title_size)
        ax.grid(True)

        # Устанавливаем шаг шкалы с шагом в 30 единиц
        xticks = np.arange(0, period + 1, 30)
        ax.set_xticks(xticks)
        plt.show()

    # 8. PACF (Частичная автокорреляционная функция)
    if plot_pacf_graph:
        fig, ax = plt.subplots(figsize=(20, 5))
        plot_pacf(df[column], ax=ax, lags=period, method='ywm')
        ax.set_title('Частичная автокорреляционная функция (PACF)', fontsize=title_size)
        ax.grid(True)

        # Устанавливаем шаг шкалы с шагом в 30 единиц
        xticks = np.arange(0, period + 1, 30)
        ax.set_xticks(xticks)
        plt.show()

    # 9. Диаграмма рассеяния с лагами
    if plot_lag_scatter:
        num_lags = len(lags)
        rows = (num_lags + 1) // 2  # Расчет количества строк для подграфиков (2 графика на строку)

        fig, axes = plt.subplots(rows, 2, figsize=(16, 6 * rows))
        axes = axes.flatten()  # Преобразуем оси в одномерный массив для универсальности

        for idx, lag in enumerate(lags):
            axes[idx].scatter(df[column], df[column].shift(lag), alpha=0.5)
            axes[idx].set_title(f'Зависимость с лагом {lag}', fontsize=title_size)
            axes[idx].set_xlabel(f'{column}(t)', fontsize=font_size)
            axes[idx].set_ylabel(f'{column}(t-{lag})', fontsize=font_size)
            axes[idx].grid(True)

        # Если количество графиков нечетное, удаляем лишние пустые оси
        if num_lags % 2 != 0:
            fig.delaxes(axes[-1])

        plt.tight_layout()
        plt.show()
```

### Оценка стационарности ряда с помощью теста Дики-Фуллера

**Цель**: Оценить, является ли временной ряд стационарным, то есть, сохраняет ли он стабильные статистические характеристики (такие как среднее и дисперсия) во времени.

**Постановка гипотез**
- **Нулевая гипотеза (H0)**: Временной ряд нестационарен (имеет тренд или изменяющуюся дисперсию).
- **Альтернативная гипотеза (H1)**: Временной ряд стационарен.

**Критерий принятия решения**
- Если p-значение теста меньше уровня значимости (alpha=0.05), нулевая гипотеза отвергается, и мы считаем, что временной ряд стационарен.
- Если p-значение больше уровня значимости, нулевая гипотеза не отклоняется, и временной ряд считается нестационарным.

```python
# Применим тест Дики-Фуллера к временному ряду
adf_test = adfuller(df['num_orders'])

# Формируем результаты теста с переводом на русский язык
adf_results_rus = {
    'Статистика ADF': adf_test[0],
    'p-значение': adf_test[1],
    'Использовано лагов': adf_test[2],
    'Количество наблюдений': adf_test[3],
    'Критические значения': adf_test[4]
}

# Вывод результатов
print("Результаты теста Дики-Фуллера:")
print(f"Статистика ADF: {adf_results_rus['Статистика ADF']:.4f}")
print(f"p-значение: {adf_results_rus['p-значение']:.4f}")
print(f"Использовано лагов: {adf_results_rus['Использовано лагов']}")
print(f"Количество наблюдений: {adf_results_rus['Количество наблюдений']}")
print("Критические значения:")
for key, value in adf_results_rus['Критические значения'].items():
    print(f"   На уровне значимости {key}: {value:.4f}")

# Интерпретация
if adf_results_rus['p-значение'] < 0.05:
    print("\nВывод: Нулевая гипотеза о нестационарности отклоняется, временной ряд стационарен на уровне значимости 5%.")
else:
    print("\nВывод: Нулевая гипотеза о нестационарности не отклоняется, временной ряд нестационарен.")
```

**Интерпретация результатов**
- Статистика ADF (-3.0689) меньше критического значения на уровне значимости 5% (-2.8622), что позволяет отклонить нулевую гипотезу на уровне значимости 5%.
- p-значение (0.0289) также меньше 0.05, что подтверждает стационарность ряда на уровне значимости 5%.

**Вывод**
Нулевая гипотеза о нестационарности отклоняется на уровне значимости 5%, следовательно, временной ряд можно считать стационарным. Это означает, что временной ряд не имеет выраженного тренда или изменяющейся дисперсии во времени, что делает его пригодным для моделирования без дополнительного преобразования.

> **Примечание**: На уровне значимости 1% этот вывод не столь однозначен, поэтому для более строгой проверки можно рассмотреть применение дополнительных методов преобразования, таких как дифференцирование.

### Проведение теста Дики-Фуллера после дифференцирования

Дифференцирование временного ряда помогает устранить тренд и сезонные колебания, делая его более стационарным. Применим разностное преобразование: каждое значение заменяется разностью между ним и предыдущим значением. Это простейшая форма дифференцирования — взятие первой разности.

```python
data= df.copy()

# Применение первой разности для устранения тренда
data['num_orders_diff'] = data['num_orders'].diff()

# Удаляем первую строку, которая теперь содержит NaN
df_diff = data.dropna(subset=['num_orders_diff'])

# Визуализируем преобразованный ряд
plt.figure(figsize=(14, 7))
plt.plot(df_diff.index, df_diff['num_orders_diff'], label='Дифференцированный временной ряд')
plt.title("Дифференцированный временной ряд")
plt.xlabel("Дата и время")
plt.ylabel("Изменение количества заказов")
plt.legend()
plt.grid(True)
plt.show()
```

```python
# Применение теста Дики-Фуллера на дифференцированном временном ряде
adf_test_diff = adfuller(df_diff['num_orders_diff'])

# Оформление результатов
adf_results_diff = {
    'Статистика ADF': adf_test_diff[0],
    'p-значение': adf_test_diff[1],
    'Критические значения': adf_test_diff[4],
    'Использовано лагов': adf_test_diff[2],
    'Количество наблюдений': adf_test_diff[3]
}

# Вывод результатов
print("Результаты теста Дики-Фуллера после дифференцирования:")
print(f"Статистика ADF: {adf_results_diff['Статистика ADF']:.4f}")
print(f"p-значение: {adf_results_diff['p-значение']:.4f}")
print("Критические значения:")
for key, value in adf_results_diff['Критические значения'].items():
    print(f"   На уровне значимости {key}: {value:.4f}")

# Интерпретация
if adf_results_diff['p-значение'] < 0.05:
    print("\nВывод: Нулевая гипотеза о нестационарности отклоняется, дифференцированный временной ряд стационарен на уровне значимости 5%.")
else:
    print("\nВывод: Нулевая гипотеза о нестационарности не отклоняется, дифференцированный временной ряд нестационарен.")
```

**Заключение**
После применения первой разности и проведения теста Дики-Фуллера результаты показывают, что дифференцированный временной ряд можно считать стационарным:

- **Статистика ADF**: -15.7932, что значительно меньше всех критических значений, особенно на уровне значимости 1% (-3.4318).
- **p-значение**: 0.0000, что указывает на сильное отклонение от нулевой гипотезы о нестационарности.

**Итоговый вывод**
Нулевая гипотеза о нестационарности отклоняется с высокой степенью уверенности на всех уровнях значимости (1%, 5% и 10%). Это означает, что после дифференцирования временной ряд стал стационарным и теперь не имеет выраженного тренда или изменяющейся дисперсии, что делает его пригодным для моделирования.

Применение дифференцирования успешно стабилизировало временной ряд, и теперь его можно использовать для анализа и построения моделей временных рядов без необходимости дополнительных преобразований для достижения стационарности.

## Обучение с использование библиотеки `ETNA`

### Функция кросс-валидации для `ETNA`

Создадим функцию (calculate_rmse_with_cv), которая будет обучать модель и выводить метрику на кросс-валидации `Timeseriessplit`

```python
def cross_validate_model_enta(
    ts_data, 
    model_class, 
    model_params, 
    pipeline_class,
    rmse_metric, 
    n_splits, 
    test_size,
    transforms=None,
    filter_outliers=False,
    apply_diff=False
):
    """
    Универсальная функция для кросс-валидации временного ряда с использованием различных моделей.

    Parameters:
    - ts_data: TSDataset — исходный временной ряд
    - model_class: класс модели для прогнозирования (например, NaiveModel, ProphetModel, CatBoost)
    - model_params: dict — параметры для инициализации модели
    - pipeline_class: класс пайплайна для обработки данных и прогнозирования
    - rmse_metric: метрика RMSE для оценки результатов
    - n_splits: количество фолдов для кросс-валидации
    - test_size: размер тестовой выборки
    - filter_outliers: bool, если True, фильтрует выбросы
    - apply_diff: bool, если True, применяет дифференцирование к временным рядам
    
    Returns:
    - mean_rmse: среднее значение RMSE по всем фолдам
    - last_valid_ts: последняя валидационная выборка (фактические значения)
    - last_forecast_ts: прогнозы на последней валидационной выборке
    """
    
    # Если выбросы нужно фильтровать, применяем фильтрацию
    if filter_outliers:
        # Извлекаем 'target' для всех сегментов
        target_values = ts_data.df.loc[:, pd.IndexSlice[:, 'target']]
        
        # Копируем данные для дальнейшей работы (будем изменять значения в этом датасете)
        ts_data_filtered = target_values.copy()

        # Проходим по всем сегментам и временным меткам
        for segment in ts_data_filtered.columns:
            for idx in ts_data_filtered.index:
                # Если текущее значение больше 400, заменяем его медианой предыдущих 720 наблюдений
                if ts_data_filtered.loc[idx, segment] > 400:
                     # Определяем диапазон индексов для окна (предыдущие 720 наблюдений)
                    current_idx = ts_data_filtered.index.get_loc(idx)
                    start_idx = max(0, current_idx - 720)
                    
                    # Находим медиану предыдущих 720 наблюдений без выбросов
                    previous_values = ts_data_filtered.iloc[start_idx:current_idx][segment]
                    median_value = np.median(previous_values[previous_values <= 400])
                    
                    # Заменяем текущее значение на медиану, если окно не пустое
                    if not np.isnan(median_value):
                        ts_data_filtered.loc[idx, segment] = median_value

        # Создаем новый TSDataset для отфильтрованных данных
        # Используем train_ts, чтобы сохранить структуру оригинальных данных (например, частоту времени и прочее)
        ts_data = TSDataset(ts_data_filtered, freq=ts_data.freq)
    
    # Применяем дифференцирование, если указано
    if apply_diff:
        ts_data_diff = ts_data.df.diff().dropna()  # первая разность
        ts_data = TSDataset(ts_data_diff, freq=ts_data.freq)   
    
    # Определяем долю данных для тестовой выборки 
    HORIZON = int(len(ts_data.df) * test_size)
    train_ts, test_ts = ts_data.train_test_split(test_size=HORIZON)

    # Кросс-валидация
    tscv = TimeSeriesSplit(n_splits=n_splits)
    
    # Список для хранения значений метрик на каждом фолде
    rmse_scores = []
    last_valid_ts = None
    last_forecast_ts = None

    # Кросс-валидация по временным рядам на основной тренировочной выборке (train_ts)
    for train_index, valid_index in tscv.split(train_ts.df):
        # Получаем метки времени для каждого фолда
        train_timestamps = train_ts.df.index[train_index]
        valid_timestamps = train_ts.df.index[valid_index]

        # Разделение данных на обучающие и валидационные фолды внутри тренировочной выборки
        fold_train_ts, valid_ts = train_ts.train_test_split(
            train_start=train_timestamps[0], train_end=train_timestamps[-1], 
            test_start=valid_timestamps[0], test_end=valid_timestamps[-1]
        )

        # Инициализация модели с параметрами
        model = model_class(**model_params)
        
        # Определяем pipeline и обучаем его
        pipeline = pipeline_class(model=model, transforms=transforms, horizon=len(valid_timestamps))
        pipeline.fit(fold_train_ts)

        # Прогноз на текущем валидационном фолде
        forecast_ts = pipeline.forecast()
        
        # Преобразуем прогноз в исходные значения, если применялось дифференцирование
        if apply_diff:
            forecast_ts_df = forecast_ts.to_pandas()  # Преобразуем в DataFrame
            last_train_value = ts_data.df.iloc[train_index[-1]].values  # Последнее известное значение
            forecast_ts_df = forecast_ts_df.cumsum() + last_train_value  # Восстанавливаем значения
            forecast_ts = TSDataset(forecast_ts_df, freq=ts_data.freq)  # Конвертируем обратно в TSDataset  

        # Оценка метрики RMSE на текущем фолде
        rmse = rmse_metric()
        rmse_score = rmse(y_true=valid_ts, y_pred=forecast_ts)
        
        # Извлекаем значение RMSE и добавляем его в список
        rmse_value = list(rmse_score.values())[0]
        rmse_scores.append(rmse_value)
        print(f'RMSE на фолде: {rmse_value}')

        # Сохраняем последнюю валидационную выборку и прогнозы
        last_valid_ts = valid_ts
        last_forecast_ts = forecast_ts
        
    # Среднее значение RMSE по всем фолдам на тренировочной выборке
    mean_rmse = np.mean(rmse_scores)
    print("Среднее значение RMSE по всем фолдам:", round(mean_rmse, 3))
    return mean_rmse, last_valid_ts, last_forecast_ts
```

### Catboost-model

Применм Catboost. Но перед этим посмотрим какие дополнительные признаки можно создать в библиотеке `ETNA`

#### Базовые transforms

##### Добавление тренда

```python
trend = LinearTrendTransform(in_column="target")
```

##### Добавление лагов

Преобразование запаздывания является наиболее простым. Оно дает нам некоторое предыдущее значение временного ряда. Например, первое запаздывание - это предыдущее значение, а пятое запаздывание - это значение, полученное пять шагов назад. Задержки необходимы для регрессионных моделей, таких как линейная регрессия или бустинг, поскольку они позволяют этим моделям получать информацию о прошлом.

Схема работы Лаг:

![lags-scheme](https://github.com/etna-team/etna/blob/master/examples/assets/get_started/lags.gif?raw=true)

Существуют некоторые ограничения на доступные задержки при прогнозировании. Мы не можем использовать предыдущее значение при составлении прогноза на последний шаг, мы просто не знаем значение. По этой причине нам следует использовать lags >= HORIZON при использовании конвейера.

```python
# Определяем долю данных для тестовой выборки
HORIZON = int(len(ts.df) * TEST_SIZE)
lags = LagTransform(in_column="target", lags=list(range(HORIZON, HORIZON + 24)), out_column="lag")
```

##### Добавления статистик

Статистика - еще одна важная функция. Она также полезна для регрессионных моделей, поскольку позволяет им анализировать информацию о прошлом, но не с помощью лагов. Существуют различные типы статистики: среднее значение, медиана, стандартное отклонение, минимум и максимум на интервале.

Схема работы Statistics:

![lags-scheme](https://github.com/etna-team/etna/blob/master/examples/assets/get_started/statistics.gif?raw=true)

Как мы видим, в окне отображается текущая временная метка. По этой причине нам не следует применять статистические преобразования к целевой переменной, мы должны применить их к целевой переменной с запаздыванием.

```python
mean = MeanTransform(in_column=f"lag_{HORIZON}", window=24)
std = StdTransform(in_column=f"lag_{HORIZON}", window=24)
quantile = QuantileTransform(in_column=f"lag_{HORIZON}", quantile=0.99, window=24)
median = MedianTransform(in_column=f"lag_{HORIZON}", window=24)
```

##### Добавения различных меток дат

Временной ряд также содержит столбец с отметкой времени, который мы еще не использовали. Но номер даты за неделю и за месяц, а также номер недели в году или флажок выходного дня могут быть действительно полезны для модели машинного обучения. И ETNA позволяет нам извлекать всю эту информацию с помощью DateFlagTransform

```python
date_flags_CatBoost = DateFlagsTransform(
    day_number_in_week=True,
    day_number_in_month=True,
    week_number_in_month=True,
    month_number_in_year=False,
    season_number=True,
    is_weekend=True,
    out_column="date_flag",
)
```

##### Добавения логарифмов

Преобразуем таргет в логарифм

```python
log = LogTransform(in_column="target", inplace=True)
```

#### Прогноз

```python
# Функция для установки случайного seed
def set_seed(seed: int = 42):
    random.seed(seed)
    np.random.seed(seed)

# Функция-цель для Optuna
def objective(
    trial: optuna.Trial, ts_data: TSDataset, n_splits: int, test_size: float, lags: int, horizon: int
):
    set_seed(42)
    
    # Определение гиперпараметров для CatBoost через Optuna
    iterations = trial.suggest_int("iterations", 50, 500)
    depth = trial.suggest_int("depth", 6, 10)
    learning_rate = trial.suggest_float("learning_rate", 1e-2, 0.1, log=True)
    window = trial.suggest_categorical("window", [24, 30, 48])
    quantile = trial.suggest_categorical("quantile", [0.99, 0.95])

    # Определение лагов
    lag_range = list(range(horizon, horizon + lags))
    
    # Определение трансформов
    transforms = [
        LagTransform(in_column="target", lags=lag_range),
        LinearTrendTransform(in_column="target"),
        DateFlagsTransform(
            is_weekend=True, 
            day_number_in_week=True, 
            day_number_in_month=True, 
            week_number_in_month=True, 
            season_number=True
        ),
        StandardScalerTransform(in_column="target"),
        MeanTransform(in_column="target", window=window),
        StdTransform(in_column="target", window=window),
        QuantileTransform(in_column="target", quantile=quantile, window=window),
        MedianTransform(in_column="target", window=window)
    ]

    # Настройка параметров модели
    model_params = {
        "iterations": iterations, 
        "depth": depth, 
        "learning_rate": learning_rate
    }

    # Используем кросс-валидацию для оценки модели с текущими гиперпараметрами
    mean_rmse, _, _ = cross_validate_model_enta(
        ts_data=ts_data,
        model_class=CatBoostPerSegmentModel,
        model_params=model_params,
        pipeline_class=Pipeline_ts,
        rmse_metric=RMSE,
        n_splits=n_splits,
        test_size=test_size,
        transforms=transforms,
        filter_outliers=True,
        apply_diff=False
    )
    
    # Возвращаем среднее значение RMSE для оптимизации
    return mean_rmse

# Основная программа для запуска Optuna
if __name__ == "__main__":
    n_trials = 3  # Количество испытаний Optuna
    n_splits = 5   # Количество фолдов для кросс-валидации
    test_size = 0.1  # Доля данных для тестовой выборки
    horizon = 24  # Горизонт прогнозирования
    lags = 48     # Количество лагов

    # Создаем исследование Optuna
    study = optuna.create_study(direction="minimize", sampler=optuna.samplers.TPESampler())

    # Запуск оптимизации
    study.optimize(
        lambda trial: objective(
            trial=trial, 
            ts_data=ts, 
            n_splits=n_splits, 
            test_size=test_size, 
            lags=lags, 
            horizon=horizon
        ),
        n_trials=n_trials
    )

    # Вывод лучших параметров
    print("Best parameters: ", study.best_params)
    print("Best value (RMSE): ", study.best_value)
```

Создадим функцию: Час, день недели и месяц являются циклическими переменными, то есть после последнего значения переменная возвращается к первому значению. Чтобы модель лучше понимала их повторяющуюся природу, можно преобразовать их с помощью синуса и косинуса.

```python
# Создадим функцию, которая создаст дополнительные признаки для модели
def make_features(data, max_lag_list, window_list):
    dataframe = data.copy()
    dataframe['hour'] = dataframe['ds'].dt.hour
    dataframe['day'] = dataframe['ds'].dt.day
    dataframe['day_of_week'] = dataframe['ds'].dt.dayofweek
    #dataframe['month'] = dataframe['ds'].dt.month
    dataframe['is_weekend'] = dataframe['day_of_week'].isin([5, 6]).astype(int)  # 5 - Суббота, 6 - Воскресенье

    # Праздники в России
    #ru_holidays = holidays.Russia()  
    #dataframe['is_holiday'] = dataframe['ds'].map(lambda x: 1 if x in ru_holidays else 0)

    # Преобразование час, день недели, месяц в синус/косинус для циклических признаков
    dataframe['sin_hour'] = np.sin(2 * np.pi * dataframe['hour'] / 24)
    dataframe['cos_hour'] = np.cos(2 * np.pi * dataframe['hour'] / 24)

    dataframe['sin_day_of_week'] = np.sin(2 * np.pi * dataframe['day_of_week'] / 7)
    dataframe['cos_day_of_week'] = np.cos(2 * np.pi * dataframe['day_of_week'] / 7)

    #dataframe['sin_month'] = np.sin(2 * np.pi * dataframe['month'] / 12)
    #dataframe['cos_month'] = np.cos(2 * np.pi * dataframe['month'] / 12)

    # Лаги для каждого значения из max_lag_list
    for max_lag in max_lag_list:
        for lag in range(1, max_lag + 1):
            dataframe[f'lag_{lag}'] = dataframe['y'].shift(lag)

    # Скользящие средние и стандартные отклонения для каждого значения из window_list
    for window in window_list:
        dataframe[f'rolling_mean_{window}'] = dataframe['y'].shift().rolling(window=window).mean()
        dataframe[f'rolling_std_{window}'] = dataframe['y'].rolling(window=window).std()

    # Применение сезонной декомпозиции к столбцу с временными метками
    decomposition = seasonal_decompose(dataframe.set_index('ds')['y'], model='additive')
    # df['trend'] = decomposition.trend.values
    dataframe['seasonal'] = decomposition.seasonal.values

    # Особые события
    #special_dates = anomaly_dates_str # Даты крупных событий
    #dataframe['is_special_event'] = dataframe.ds.map(lambda x: 1 if str(x.date()) in special_dates else 0)
    
    return dataframe    
```

```python
# Список для хранения метрик на каждом фолде
rmse_scores_prophet = []

# Кросс-валидация по временным рядам
tscv = TimeSeriesSplit(n_splits=N_SPLITS)

for train_index, val_index in tscv.split(prophet_data):
    X_train_prophet, X_val_prophet = X_prophet.iloc[train_index], X_prophet.iloc[val_index]
    y_train_prophet, y_val_prophet = y_prophet.iloc[train_index], y_prophet.iloc[val_index]

    # Подготовка данных для Prophet
    prophet_train_data = X_train_prophet[['ds']].copy()
    prophet_train_data['y'] = y_train_prophet.copy()
    prophet_val_data = X_val_prophet[['ds']].copy()

    # Инициализация и настройка модели Prophet
    model_prophet_sklearn = Prophet(yearly_seasonality=False, weekly_seasonality=True, daily_seasonality=True)

    # Добавляем регрессоры
    regressors_prophet_sklearn = ['hour', 'day', 'day_of_week', 'is_weekend', 'sin_hour', 'cos_hour',
                                  'sin_day_of_week', 'cos_day_of_week', 'seasonal'] + \
                                 [f'lag_{lag}' for lag in range(1, 9)] + \
                                 [f'rolling_mean_{window}' for window in [7, 24]] + \
                                 [f'rolling_std_{window}' for window in [7, 24]]

    for regressor in regressors_prophet_sklearn:
        model_prophet_sklearn.add_regressor(regressor)

    # Обучение модели Prophet
    model_prophet_sklearn.fit(pd.concat([prophet_train_data, X_train_prophet[regressors_prophet_sklearn]], axis=1))

    # Создание датафрейма с регрессорами для валидационной выборки
    future_prophet_val = pd.concat([prophet_val_data, X_val_prophet[regressors_prophet_sklearn]], axis=1)

    # Прогнозирование
    forecast_prophet_val = model_prophet_sklearn.predict(future_prophet_val)
    predictions_prophet_val = forecast_prophet_val['yhat']

    # Оценка модели (RMSE) на валидационной выборке
    rmse_val_prophet = sqrt(mean_squared_error(y_val_prophet, predictions_prophet_val))
    rmse_scores_prophet.append(rmse_val_prophet)
    print(f"RMSE на фолде: {rmse_val_prophet}")

# Среднее значение RMSE по всем фолдам
mean_rmse_prophet = sum(rmse_scores_prophet) / len(rmse_scores_prophet)
print("Среднее значение RMSE на валидационной выборке для Prophet:", mean_rmse_prophet)
```

### Создание пайплайна

```python
data.set_index('ds', inplace=True)

# Разбиение данных на признаки и целевую переменную 
X = data.drop(columns=['y'], axis=1).copy()
y = data['y'].copy()

# Разбиение на обучающую и тестовую выборки   
X_train, X_test, y_train, y_test = train_test_split(
    X, y, shuffle=False, test_size=TEST_SIZE_1, random_state=RANDOM_STATE
)

print(X_train.shape, X_test.shape)
print(LINE_1)
y_train.shape, y_test.shape
```

```python
# создаём списки с названиями признаков
cat_columns = ['is_weekend']
num_columns = [col for col in X.columns if col not in cat_columns]

print("Категориальные столбцы:", cat_columns)
print("Числовые столбцы:", num_columns)
```

```python
# Определение двух пайплайнов для кодирования
def get_preprocessor(encoder_type):
    if encoder_type == 'ohe':
        encoder = OneHotEncoder(drop='first', sparse_output=False, handle_unknown='ignore')
    elif encoder_type == 'label':
        encoder = OrdinalEncoder(handle_unknown='use_encoded_value', unknown_value=-1)
    else:
        raise ValueError(f"Неизвестный тип кодировщика: {encoder_type}")
    
    preprocessor = ColumnTransformer(
        transformers=[
            ('cat', Pipeline(steps=[
                ('imputer', SimpleImputer(strategy='most_frequent')),
                ('encoder', encoder)
            ]), cat_columns),
            ('num', MinMaxScaler(), num_columns)
        ],
        remainder='passthrough'
    )

    return preprocessor



# Обновляем функцию для создания пайплайна с полиномиальными признаками
def pipe_line(cat_columns, num_columns, model):
    if isinstance(model, (LinearRegression, Ridge, Lasso, ElasticNet)):
        preprocessor = get_preprocessor('ohe')
        # Добавляем шаг с полиномиальными признаками для линейных моделей
        pipe_final = Pipeline([
            ('preprocessor', preprocessor),
            ('poly', PolynomialFeatures(degree=2, include_bias=False)),  # Шаг полиномиальных признаков
            ('models', model)
        ])
    else:
        preprocessor = get_preprocessor('ord')
        pipe_final = Pipeline([
            ('preprocessor', preprocessor),
            ('models', model)
        ])
    return pipe_final
```

### Catboost

#### Кроссс-валидация

```python
# Сохранение прогнозов и фактических значений последнего фолда
last_fold_predictions = None
last_fold_y_true = None

def objective_catboost_cv(trial):
    global last_fold_predictions, last_fold_y_true
    
    # Определение гиперпараметров
    param_catboost = {
        'iterations': trial.suggest_int('iterations', 200, 600),
        'learning_rate': trial.suggest_loguniform('learning_rate', 0.01, 0.1),
        'depth': trial.suggest_int('depth', 4, 6),
        'l2_leaf_reg': trial.suggest_loguniform('l2_leaf_reg', 2, 3),
        'loss_function': 'RMSE',
        'eval_metric': 'RMSE',
        'random_seed': RANDOM_STATE,
        'early_stopping_rounds': 10
    }
    
    # Создание модели
    model = CatBoostRegressor(**param_catboost)
    
    # Использование TimeSeriesSplit для кросс-валидации
    tscv = TimeSeriesSplit(n_splits=n_splits)
    rmse_scores = []
    
    # Проход по фолдам
    for fold_idx, (train_idx, valid_idx) in enumerate(tscv.split(X_train)):
        X_fold_train, X_fold_valid = X_train.iloc[train_idx], X_train.iloc[valid_idx]
        y_fold_train, y_fold_valid = y_train.iloc[train_idx], y_train.iloc[valid_idx]
        
        # Обучение модели на текущем фолде
        model.fit(X_fold_train, y_fold_train, cat_features=cat_columns, verbose=False)
        
        # Прогнозирование на валидационном фолде
        y_pred_fold = model.predict(X_fold_valid)
        rmse_fold = mean_squared_error(y_fold_valid, y_pred_fold, squared=False)
        rmse_scores.append(rmse_fold)
        
        # Сохранение прогнозов и фактических значений для последнего фолда
        if fold_idx == n_splits - 1:
            last_fold_predictions = y_pred_fold
            last_fold_y_true = y_fold_valid
    
    # Возвращаем среднее значение RMSE по всем фолдам
    return np.mean(rmse_scores)
```

```python
# Запуск Optuna для поиска лучших гиперпараметров
study_catboost = optuna.create_study(direction='minimize', sampler=optuna.samplers.TPESampler(seed=RANDOM_STATE))
study_catboost.optimize(objective_catboost_cv, n_trials=10)

# Вывод лучших результатов
print('Лучшая модель:')
best_trial_catboost = study_catboost.best_trial
best_score_catboost = best_trial_catboost.value
print(f'Средний RMSE на валидации по фолдам: {best_score_catboost}')
print('Параметры модели:')
for key, value in best_trial_catboost.params.items():
    print(f'    {key}: {value}')
```

## Анализ важность признаков

```python
# Получаем названия признаков после всех преобразований
preprocessor = best_model_linear.named_steps['preprocessor']

# Извлекаем названия категориальных и числовых признаков
cat_features = preprocessor.transformers_[0][1].named_steps['encoder'].get_feature_names_out(['is_weekend'])
num_features = np.array([
    'hour', 'day', 'day_of_week', 'sin_hour', 'cos_hour', 
    'sin_day_of_week', 'cos_day_of_week', 
    'lag_1', 'lag_2', 'lag_3', 'lag_4', 'lag_5', 'lag_6', 'lag_7', 'lag_8',
    'rolling_mean_7', 'rolling_std_7', 'rolling_mean_24', 'rolling_std_24', 'seasonal'
])

# Комбинируем числовые и категориальные признаки
initial_feature_names = np.hstack([cat_features, num_features])

# После применения PolynomialFeatures названия признаков увеличатся
poly = best_model_linear.named_steps['poly']
expanded_feature_names = poly.get_feature_names_out(initial_feature_names)

# Получаем коэффициенты модели
coefficients = best_model_linear.named_steps['models'].coef_

# Создаем DataFrame для удобства визуализации
importance_df = pd.DataFrame({
    'feature': expanded_feature_names,
    'importance': coefficients
})

# Сортируем признаки по важности
importance_df = importance_df.sort_values(by='importance', ascending=False)

# Получаем топ-20 положительных и топ-20 отрицательных признаков
top_positive = importance_df.head(20)
top_negative = importance_df.tail(20)

# Объединяем их для отображения на одном графике
top_features = pd.concat([top_positive, top_negative])

# Визуализация топ-20 положительных и топ-20 отрицательных признаков
plt.figure(figsize=(15, 10))
plt.barh(top_features['feature'], top_features['importance'], color='skyblue')
plt.xlabel('Важность признаков (коэффициенты)')
plt.title('Топ-20 положительных и топ-20 отрицательных признаков для модели ElasticNet')
plt.grid(True)
plt.show()
```

**ВЫВОД:**
Из представленного графика видно, что у нас имеются как положительные, так и отрицательные коэффициенты важности признаков для модели ElasticNet (Лучшая модель). Положительные значения указывают на признаки, которые увеличивают зависимую переменную, тогда как отрицательные значения указывают на признаки, которые её уменьшают. Наибольшее положительное влияние оказывают признаки, связанные со скользящим средним и лагами. Отрицательные коэффициенты также связаны с временными признаками, но их влияние ниже.

Наибольшую важность имеют признаки, связанные с временными сдвигами (lag) и скользящими средними (rolling_mean, rolling_std), что типично для временных рядов, где прошлые значения могут быть хорошими предикторами для будущих.
